```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# -------- CASE 1 : Logistic distribution --------
alpha = 0     # example mean
beta = 1      # example scale

def simulate_logistic(alpha, beta, n=1000):
    U = np.random.rand(n)
    X = alpha - beta * np.log(1/U - 1)
    return X

X1 = simulate_logistic(alpha, beta, 1000)
```

```python
# -------- CASE 2 : Semicircle distribution --------
R = 1  # example radius

def simulate_semicircle(R, n=100):
    theta = np.random.uniform(0, np.pi, n)
    return R * np.cos(theta)

X2 = simulate_semicircle(R, 100)
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.hist(X1, bins=20, density=True, color="skyblue", edgecolor="black")
plt.title("Logistic distribution simulation")

plt.subplot(1,2,2)
plt.hist(X2, bins=20, density=True, color="salmon", edgecolor="black")
plt.title("Semicircle distribution simulation")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# --------------------
# Parameters
# --------------------
alpha = 0
beta = 1
R = 1

# --------------------
# 1. Logistic distribution
# --------------------
def simulate_logistic(alpha, beta, n=1000):
    U = np.random.rand(n)
    X = alpha + beta * np.log(U / (1 - U))
    return X

log_X = simulate_logistic(alpha, beta)

# --------------------
# 2. Disk marginal distribution
# --------------------
def simulate_disk_x(R, n=1000):
    U1 = np.random.rand(n)
    U2 = np.random.rand(n)

    r = R * np.sqrt(U1)
    theta = 2 * np.pi * U2

    X = r * np.cos(theta)
    return X

disk_X = simulate_disk_x(R)

# --------------------
# Plotting
# --------------------
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.hist(log_X, bins=15, density=True, color='skyblue', edgecolor='black')
plt.title("Logistic Distribution Samples")

plt.subplot(1,2,2)
plt.hist(disk_X, bins=15, density=True, color='salmon', edgecolor='black')
plt.title("Disk Marginal Distribution Samples")

plt.tight_layout()
plt.show()
```
